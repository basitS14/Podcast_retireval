```python
nlp = spacy.load("en_core_web_sm")
stop_words = stopwords.words('english')

def preprocessing(df , columns):
    for cols in columns:
        df[cols] = df[cols].apply(lambda x : str(x))
        df[cols] = df[cols].apply(lambda x : x.lower())
        df[cols] = df[cols].apply(lambda x :[token.text for token in nlp(x) if not token.is_stop])
        
    return df
```
